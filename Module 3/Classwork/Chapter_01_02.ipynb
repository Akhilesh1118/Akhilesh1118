# Pie Chart
- Data Distribution

```python
#import libraries
import matplotlib.pyplot as plt
```

```python
# Quarter names
quarters = ["Q1", "Q2", "Q3", "Q4"]

# Colors for each quarter
colors = ["pink", "red", "green", "brown"]

# Profit values for each quarter
profit = [40_000, 50_000, 50_000, 120_000]

# Set the chart size
plt.figure(figsize=(4, 3))

# Create a pie chart
plt.pie(
    profit,
    labels=quarters,              # Display quarter labels
    autopct="%1.0f%%",            # Show percentage values on slices
    startangle=40,                # Rotate the chart by 40 degrees
    explode=(0.1, 0.1, 0.1, 0.1), # Separate each slice slightly
    shadow=True,                  # Add a shadow effect
    colors=colors                 # Apply custom colors
)

# Keep the pie chart circular
plt.axis("equal")

# Add a chart title
plt.title("Profit vs Quarters")

# Display the chart
plt.show()
```

# Scatter plots

```python
# Define data
Hours = [2,3,4,6,8]
Income = [1000,1500,2500,2500,5000]
size = [100, 1000,3000,4000, 2400]

# Colors for each quarter
colors = ["pink", "red", "green", "brown", "blue"]

# Plot the scatter chart
plt.scatter(Hours,Income,s=size,c=colors, alpha=0.7)

plt.colorbar(label="Intensity")
# Add a chart title
plt.title("Hours vs Income")

#  Add label x and y
plt.xlabel("Hours")
plt.ylabel("Income")

# plot the graph
plt.show()
```

```python
# Number of hours worked
hours = [2, 3, 4, 6, 8]

# Income earned for each hour value
income = [1000, 1500, 2500, 2500, 5000]

# Size of each point in the scatter plot
sizes = [100, 1000, 3000, 4000, 2400]

# Color for each point
colors = ["pink", "red", "green", "brown", "blue"]

# Create a scatter plot
plt.scatter(
    hours,
    income,
    s=sizes,      # Set the size of each point
    c=colors,     # Set the color of each point
    alpha=0.7     # Make points slightly transparent
)

# Add a color scale (works best when colors are numeric values)
plt.colorbar(label="Intensity")

# Add a title to the chart
plt.title("Hours vs Income")

# Add labels for X-axis and Y-axis
plt.xlabel("Hours Worked")
plt.ylabel("Income Earned")

# Display the chart
plt.show()
```

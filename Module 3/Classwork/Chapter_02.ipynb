## Advance Data Visualization with Seaborn

# Seaborn
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics with minimal code.

## Seaborn provides built-in themes:
1. Darkgrid
2. Whitegrid
3. Dark  
4. White
5. Ticks

Also supports :


```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Load

```python
df = sns.load_dataset("penguins")
df
```

```python
df.columns
```

```python
df.info()
```

```python
df2 = df.head(15) # Shows the first 5 rows .
```

# Line Chart

```python
sns.lineplot(
    x="bill_length_mm",
    y="bill_depth_mm",
    # data=df.head(15)
    data=df2 ,
    hue='sex' ,
    legend=True ,
    style='species',
    markers=["o" , "^"] ,
    size='sex' ,
    sizes=(2,5),
    palette='cubehelix'
)
```

# Apply Theme

```python
sns.set_theme(style="whitegrid")

sns.lineplot(
    x="bill_length_mm",
    y="bill_depth_mm",
    # data=df.head(15)
    data=df2 ,
    hue='sex' ,
    legend=True ,
    style='species',
    markers=["o" , "^"] ,
    size='sex' ,
    sizes=(2,5),
    palette='cubehelix'
)
```

```python
sns.set_theme(style="ticks")


sns.lineplot(
    x="bill_length_mm",
    y="bill_depth_mm",
    # data=df.head(15)
    data=df2 ,
    hue='species' ,
    legend=True ,
    style='species',
    markers=["o" , "^"] ,
    size='species' ,
    sizes=(5,5),
    palette='cubehelix'
)
```

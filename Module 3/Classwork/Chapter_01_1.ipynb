# Histogram
What is a Histogram?

A histogram is a graphical representation of the distribution of numerical\
data. It groups data into intervals called bins and shows how frequently\
 values occur.

### Why Use a Histogram?
* To understand data distribution.
* To identify patterns and trends.
* To  detect outliers.
* To analyze the spread of data.

**Important Terms**
* Bins: Intervals used to group data.
* Frequency: Number of observations in each bin.
* Distribution: Overall pattern of the data.

```python
# import libraries.
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create sample data
marks = [50,56,72,88,98,60,62,74,92,90]

# plot graph
plt.hist(marks)
plt.show()
```

```python
from matplotlib.lines import lineStyles
# Create sample data
marks = [50,56,72,88,98,60,62,74,92,90]

# plot graph
plt.hist(marks)
plt.hist(marks , bins = 5 , edgecolor='brown' , color='beige' , alpha = 0.9)

# give title
plt.title("marks distribution")

# add labels and title in charts
plt.xlabel("Marks")
plt.ylabel("Frequency")

# add grids
plt.grid(axis="y" , color='tan' , linestyle='--' )

plt.show()
```

```python

# Create sample data
marks = [10,20,30,40,50,60,70,80]

#adjust figure
plt.figure(figsize=(8,5))

# plot graph
plt.hist(marks)
plt.hist(marks , bins = 5 , edgecolor='brown' , color='beige' , alpha = 0.9)

# give title
plt.title("marks distribution")

# add labels and title in charts
plt.xlabel("Marks")
plt.ylabel("Frequency")

# add grids
plt.grid(axis="y" , color='tan' , linestyle='--' )

plt.show()
```

```python
# Data set
marks = [10, 20, 30, 40, 50, 60, 70, 80]

# Calculate median
median_mark = np.median(marks)

# Adjust figure
plt.figure(figsize=(8, 5))

# Plot histogram
plt.hist(
    marks,
    bins=5,
    edgecolor='brown',
    color='beige',
    alpha=0.9
)

# Add median line
plt.axvline(
    median_mark,
    color='red',
    linestyle='--',
    linewidth=2,
    label=f'Median = {median_mark}'
)

# Give title
plt.title("Marks Distribution")

# Add labels
plt.xlabel("Marks")
plt.ylabel("Frequency")

# Add grid
plt.grid(axis="y", color='tan', linestyle='--')

# Show legend
plt.legend()

plt.show()
```

# Bar Chart - Category comparison

2. Grouped Bar Chart

```python
# Create sample data
subject = ["Python" , "C", "java" , "C++"]
marks = [50,56,72,88]


#adjust figure
plt.figure(figsize=(8,5))

# plot graph
plt.bar(subject , marks , color='tan'   )

# give title
plt.title("Score")

# add labels and title in charts
plt.xlabel("subject")
plt.ylabel("Marks")

# add the value in the top of bar
for index, value in enumerate(marks):
    plt.text(index, value+1, str(value), ha='center')
plt.show()
```
